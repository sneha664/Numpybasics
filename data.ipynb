```python

import numpy as np
import pandas as pd

data = np.array([10,12,12,13,12,14,15,16,100,200])
df = pd.DataFrame(data, columns=['Value'])

print("Original data:")
print(df)

def detect_outliers_iqr(data):
    Q1 = data['Value'].quantile(0.25)
    Q3 = data['Value'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return (data['Value'] < lower_bound) | (data['Value'] > upper_bound)

outliers = detect_outliers_iqr(df)

print("\nDetected outliers:")
print(df[outliers])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = np.array([10,12,12,13,12,14,15,16,100,200])
df = pd.DataFrame(data, columns=['Value'])

print("Original data:")
print(df)

def detect_outliers_iqr(data):
    Q1 = data['Value'].quantile(0.25)
    Q3 = data['Value'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return (data['Value'] < lower_bound) | (data['Value'] > upper_bound)

outliers = detect_outliers_iqr(df)

print("\nDetected outliers:")
print(df[outliers])

plt.figure(figsize=(15,6))
plt.scatter(range(len(df)), df['Value'], 
            c=['red' if x else 'blue' for x in outliers], 
            label='Data points')

plt.axhline(y=df['Value'].mean(), color='green', linestyle='--', label='Mean')
plt.title('Outlier Detection')
plt.xlabel('Index')
plt.ylabel('Data Value')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df['Value'])
plt.title('Dataset with outliers (Box Plot)')
plt.show()
```
