```python
import numpy as np
import pandas as pd
```

```python
data={
    'ID':[1,2,3,4,5,6],
    'Name':['Aarya','Aarav','Priya','Pooja','Shivam','Rashi'],
    'location':['Kathmandu','Biratnagar','Pokhara','Pokhara','Kathmandu','Chitwan'],
    'Salary':[20000,30000,20000,41000,None,None],
    'Age':[20,21,22,45,24,25]}
df=pd.DataFrame(data)
print(df)
    
    
```

```python
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print("\nAfter Filling NaN With Mean:")
print(df)

```

```python
median_salary =df['Salary'].median()
df['Salary']=df['Salary'].fillna(median_salary)
print("\n filling NaN with median:")
df
```

```python

data={
    'ID':[1,2,3,4,5,6],
    'Name':['Aarya','Aarav','Priya','Pooja','Shivam','Rashi'],
    'location':['Kathmandu','Biratnagar','Pokhara','Pokhara','Kathmandu','Chitwan'],
    'Salary':[20000,30000,20000,41000,None,None],
    'Age':[20,21,22,45,24,25]}
df=pd.DataFrame(data)

df['Salary'] = df['Salary'].fillna(df['Salary'].mode()[0])

print(df)

```

```python
data={
    'ID':[1,2,3,4,5,6],
    'Name':['Aarya','Aarav','Priya','Pooja','Shivam','Rashi'],
    'location':['Kathmandu','Biratnagar','Pokhara','Pokhara','Kathmandu','Chitwan'],
    'Salary':[20000,30000,20000,41000,None,None],
    'Age':[20,21,22,45,24,25]}
df=pd.DataFrame(data)
df['Salary'] = df['Salary'].fillna(df['Salary'].mode()[0])
age_median = df['Age'].median()
df.loc[df['Age'] == 45, 'Age'] = age_median
print(df)
```

```python
 import matplotlib.pyplot as plt
data = {
    'ID':[1,2,3,4,5,6],
    'Name':['Aarya','Aarav','Priya','Pooja','Shivam','Rashi'],
    'location':['Kathmandu','Biratnagar','Pokhara','Pokhara','Kathmandu','Chitwan'],
    'Salary':[20000,30000,20000,41000,20000,20000],
    'Age':[20,21,22,23,24,25]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Salary'], color='skyblue')
plt.title('Salary of Employees')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
avg_salary = df.groupby('location')['Salary'].mean().reset_index()
plt.figure(figsize=(8,5))
plt.bar(avg_salary['location'], avg_salary['Salary'], color='orange')
plt.title('Average Salary by Location')
plt.xlabel('Location')
plt.ylabel('Average Salary')
plt.show()
```
